```python
def initialize_agents():
    llm = creat_llm()

    manage_agent = ManageAgent(llm)
    clue_agent = ClueAgent(llm)
    feedback_agent = FeedbackAgent(llm)

    npc_agents = {npc_info['角色姓名']: NPCAgent(llm=llm, name=npc_info['角色姓名'], info=npc_info) for npc_info in npc_characters}

    user_agent = UserAgent(llm, scenario_background['用户角色信息'])
    return manage_agent, clue_agent, feedback_agent, npc_agents, user_agent


def main():
    dialog_history_memery = []
    manage_agent, clue_agent, feedback_agent, npc_agents, user_agent = initialize_agents()
    npc_agent_name_list = list(npc_agents.keys())
    npc_statuses = {npc['角色姓名']: npc['初始心情值'] for npc in npc_characters}
    for i in range(20):
        # Manage Agent for Conversation Management
        manage_agent_output = manage_agent.generate_response(input_dict={
            "scenario_background": scenario_background,
            "dialog_history": dialog_history_memery,
            "npc_statuses": npc_characters
        })
        goal_completion_rate = manage_agent_output.get("goal_completion_rate", "0%")
        if goal_completion_rate == "100%":
            print("目标已完全达成，对话结束。")
            break

        plot_direction = manage_agent_output.get("plot_direction", "")

        next_speaker = manage_agent_output['next_speaker']
        if next_speaker == "用户":
            # user_input = input("请输入用户的发言：")
            user_input = user_agent.generate_response(input_dict={
                "scenario_background": scenario_background,
                "dialog_history": dialog_history_memery,
                "plot_direction": plot_direction
            })['response']
            dialog_history_memery.append({"speaker": "用户", "content": user_input})
            print(f"{dialog_history_memery[-1]['speaker']}: {dialog_history_memery[-1]['content']}")

            feedback = feedback_agent.generate_response(input_dict={
                "npc_characters": npc_characters,
                "npc_statuses": npc_statuses,
                "user_input": user_input,
                "dialog_history": dialog_history_memery
            })
            mood_changes = feedback.get("mood_change", [])
            for change in mood_changes:
                npc_name = change["npc"]
                mood_value_change = int(change["mood_change"])
                npc_statuses[npc_name] += mood_value_change
            dialog_history_memery.append({"NPCs实时心情值": npc_statuses})
            print(f"NPCs实时心情值: {npc_statuses}")

        elif next_speaker in npc_agent_name_list:
            npc_agent = npc_agents[next_speaker]
            npc_agent_output = npc_agent.generate_response(input_dict={
                "scenario_background": scenario_background,
                "dialog_history": dialog_history_memery,
                "plot_direction": plot_direction
            })
            dialog_history_memery.append({"speaker": next_speaker, "content": npc_agent_output['response']})

            print(f"{dialog_history_memery[-1]['speaker']}: {dialog_history_memery[-1]['content']}")
   

if __name__ == "__main__":
    main()
```
